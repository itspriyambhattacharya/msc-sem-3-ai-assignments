```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('./datasets/golf_dataset_the_origin.csv')
```

```python
y = df['Play']
X = df.drop('Play', axis=1)
```

```python
# label encoding
le = LabelEncoder()

X['Outlook'] = le.fit_transform(X['Outlook'])
```

```python
model = DecisionTreeClassifier(
    criterion='entropy'
)

model.fit(X, y)
```

```python
plt.figure(figsize=(29, 19))
plot_tree(
    model,
    filled=True,
    feature_names=X.columns,
    class_names=['1', '0']
)
plt.show()
```

```python
model2 = GaussianNB()
model2.fit(X, y)
```

```python
newSample = pd.DataFrame(
    [['sunny', 76, 95, 1]],
    columns=X.columns
)
```

```python
newSample['Outlook'] = le.transform(newSample['Outlook'])

pred = model2.predict(newSample)
print(pred[0])
```
