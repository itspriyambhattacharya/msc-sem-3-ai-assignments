```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import skfuzzy as fuzz
```

```python
df = pd.read_csv('./datasets/mtcars.csv')
df = df.drop('model', axis =1)
X = df.to_numpy()
Xt = X.astype(np.float64).T # transposed X
c = 2 # number of clusters
```

```python
ctr, u,_, _, _, _ ,fpr = fuzz.cluster.cmeans(
    Xt, c, m=2, error=0.005, maxiter = 2000
)

print(u)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(ctr[:, 0], ctr[:, 1], marker='X')
plt.title("Fuzzy C-Means Clustering")
plt.show()
```
